# Scipy - обрезка изображения
## Выведите изображения:
## 1) Цветовая карта;
## 2) Обрезанное изображение;
## 3) Изображение, где лицо выделено при помощи кружка.


```python
#Вывод оригинального изображеня
import numpy as np
from scipy import misc
import matplotlib.pyplot as plt

face = misc.face(gray=True)
plt.imshow(face)
```

Цветовая карта:

```python
plt.imshow(face, cmap=plt.cm.gray)
```

Обрезка изображения:

```python
from PIL import Image
face_array = np.array(face)
crop_face = face[100:-100, 100:-100]
cropped_face = Image.fromarray(crop_face)
plt.imshow(cropped_face, cmap=plt.cm.gray)
```

Обрезка лица(круг):

```python
sy, sx = face.shape
y, x = np.ogrid[0:sy, 0:sx]
y.shape, x.shape

centerx, centery = (660, 300)
mask = ((y - centery)**2 + (x - centerx)**2) > 230**2 # круг

face[mask] = 0
plt.imshow(face, cmap=plt.cm.gray) 
```

Обрезка лица(овал):

```python
sy, sx = face.shape
y, x = np.ogrid[0:sy, 0:sx]
y.shape, x.shape

centerx, centery = (660, 340)  # Центр овала
radius_x = 220  # Радиус по оси X
radius_y = 150  # Радиус по оси Y

mask = (((y - centery) ** 2) / radius_y ** 2) + (((x - centerx) ** 2) / radius_x ** 2) > 1 #овал
face[mask] = 0
plt.imshow(face, cmap=plt.cm.gray) 
```
