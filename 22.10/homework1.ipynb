## Подсчёт числа PI методом Монте-Карло

Необходимые формулы:
1) Площадь круга: $c = \pi r^2$
2) Площадь квадрата: $s = (2 r)^2$
3) $c/s = (\pi r^2)  / (4 r^2) =   \pi / 4$  
4) $ \pi = 4 * c/s$ 

Изображение для пояснения концепции:



Импорт используемых модуей:

```python
import random
import matplotlib.pyplot as plt 
```

Количество точек:

```python
num_points = 10000
```

Механизм броска:

```python
points = []
hits = 0
for _ in range (num_points):
    x,y = random.random(), random.random() 
    if x*x + y*y < 1.0:
        hits += 1 
        points.append((x,y,"red")) 
    else:
        points.append((x,y,"blue"))
```

Построение графика результатов и оценка числа $\pi$:

```python
x, y, colors = zip(*points)
fig, ax = plt.subplots() 
fig.set_size_inches(6.0, 6.0)
ax.scatter(x, y, c=colors)
fraction = hits / num_points
print("pi = ", 4 * fraction)
```
