# Выгрузить страницу https://en.wikipedia.org/wiki/Python и сформировать список ссылок, которые есть на этой странице.

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
response = urlopen('https://en.wikipedia.org/wiki/Python')
html = response.read()
soup = BeautifulSoup(html, 'html.parser')
links = soup.find_all('a')
#print(links)
link_list = [link.get('href') for link in links 
             if link.get('href') is not None]
        
for link in link_list:
    print(link)
```
